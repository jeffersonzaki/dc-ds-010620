Read data from `./data/weatherAUS.csv` and assign it to a variable named **`weather_df`**.

1. Fill all `Sunshine` Nans with **`mean_sunshine`**

````python
weather_df = pd.read_csv("weatherAUS.csv")
````

````python
# Creating a datetime format to the Date column
weather_df["Date"] = pd.to_datetime(weather_df["Date"])

# Adding new columns with values from the Date column
weather_df["day"] = weather_df["Date"].dt.day
weather_df["month"] = weather_df["Date"].dt.month
weather_df["year"] = weather_df["Date"].dt.year
````

````python
# df.columns[df.isnull().any()]
incomplete_columns = list(weather_df.columns[weather_df.isna().any()])
````

````python
# Dropping columns
def keep_cols(DataFrame, keep_these):
    """
    Keep only the columns [keep_these] in a DataFrame, delete
    all other columns. 
    """
    drop_these = list(set(list(DataFrame)) - set(keep_these))
    return DataFrame.drop(drop_these, axis = 1)


keep_cols(weather_df, ["day", "month", "year", "Rainfall", "Sunshine"])
````

````python
# Dropping missing rows from Sunshine
weather_df.dropna(how="any", subset=["Sunshine"], inplace=True)
````

````python
mean_sunshine = weather_df.Sunshine.mean()
````

````python
%matplotlib inline


fig, ax = plt.subplots()

# Example data
ylabel = ("A", "B", "C", "D", "E")
xlabel = (54, 23, 99, 37, 3)

y_pos = np.arange(len(ylabel))
x_pos = np.arange(len(xlabel))

# Creating graph
ax.barh(y_pos, x_pos, color="red")

# Labeling
ax.set_yticks(y_pos)
ax.set_xticks(x_pos)

ax.set_yticklabels(ylabel)
ax.set_yticklabels(xlabel)

ax.invert_yaxis()  # labels read top-to-bottom

ax.set_ylabel("Category")
ax.set_xlabel("Count")
ax.set_title("Count By Category")

plt.show()
````
